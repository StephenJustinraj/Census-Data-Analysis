```python
population_2001['State'] = population_2001['State'].replace('Delhi', 'Nct')
```

```python
population_2001 = population_2001.sort_values(by='State').reset_index(drop=True)
```

```python
population_2001.iloc[[5, 6]] = population_2001.iloc[[6, 5]].values
population_2001.iloc[[7, 8]] = population_2001.iloc[[8, 7]].values
population_2001.iloc[[11, 12]] = population_2001.iloc[[12, 11]].values
```

```python
merged_pop = pd.DataFrame({
    'State': population_2011['State name'],  # Column from df1
    'State2': population_2001['State'],  # Column from df1
    'Population_2001': population_2001['Persons'],  # Column from df1
    'Population_2011': population_2011['Population']   # Column from df2
})
```

```python
merged_pop
```

```python
merged_pop['Population Change Percentage'] = ((merged_pop['Population_2011'] - merged_pop['Population_2001']) / merged_pop['Population_2001']) * 100

# Sort the states based on the highest percentage change
top_5_population_increase = merged_pop.sort_values(by='Population Change Percentage', ascending=False).head(5)

# Plot the results
plt.figure(figsize=(10,6))
plt.bar(top_5_population_increase['State'], top_5_population_increase['Population Change Percentage'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Population Change Percentage')
plt.title('Top 5 States with Highest Population Increase Percentage (2001 to 2011)')
plt.xticks(rotation=45)
plt.show()
```

```python
merged_pop['Population Change'] = merged_pop['Population_2011'] - merged_pop['Population_2001']

# Sort the dataframe by Population Change in descending order
merged_pop_sorted = merged_pop.sort_values(by='Population Change', ascending=False)

# Select the top 5 states with the highest population increase
top_5_states = merged_pop_sorted.head(5)

# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(top_5_states['State'], top_5_states['Population Change'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Population Increase')
plt.title('Top 5 States with the Highest Population Increase (2001 to 2011)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_2011['Gender_Ratio'] = df_2011['Male'] / df_2011['Female']
df_2011['Literacy_Rate'] = (df_2011['Literate'] / df_2011['Population']) * 100
df_2011['Male_Literacy'] = (df_2011['Male_Literate'] / df_2011['Male']) * 100
df_2011['Female_Literacy'] = (df_2011['Female_Literate'] / df_2011['Female']) * 100

# Get top and bottom 5 for each metric
top_5_gender_ratio = df_2011.nlargest(5, 'Gender_Ratio')
bottom_5_gender_ratio = df_2011.nsmallest(5, 'Gender_Ratio')

top_5_literacy_rate = df_2011.nlargest(5, 'Literacy_Rate')
bottom_5_literacy_rate = df_2011.nsmallest(5, 'Literacy_Rate')

top_5_male_literacy = df_2011.nlargest(5, 'Male_Literacy')
bottom_5_male_literacy = df_2011.nsmallest(5, 'Male_Literacy')

top_5_female_literacy = df_2011.nlargest(5, 'Female_Literacy')
bottom_5_female_literacy = df_2011.nsmallest(5, 'Female_Literacy')

# Plotting the top and bottom 5 for each metric

fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Gender Ratio Plot
axs[0, 0].barh(top_5_gender_ratio['District name'], top_5_gender_ratio['Gender_Ratio'], color='green', label='Top 5')
axs[0, 0].barh(bottom_5_gender_ratio['District name'], bottom_5_gender_ratio['Gender_Ratio'], color='red', label='Bottom 5')
axs[0, 0].set_title('Top and Bottom 5 Gender Ratios')
axs[0, 0].set_xlabel('Gender Ratio')
axs[0, 0].legend()

# Literacy Rate Plot
axs[0, 1].barh(top_5_literacy_rate['District name'], top_5_literacy_rate['Literacy_Rate'], color='blue', label='Top 5')
axs[0, 1].barh(bottom_5_literacy_rate['District name'], bottom_5_literacy_rate['Literacy_Rate'], color='orange', label='Bottom 5')
axs[0, 1].set_title('Top and Bottom 5 Literacy Rates')
axs[0, 1].set_xlabel('Literacy Rate (%)')
axs[0, 1].legend()

# Male Literacy Plot
axs[1, 0].barh(top_5_male_literacy['District name'], top_5_male_literacy['Male_Literacy'], color='purple', label='Top 5')
axs[1, 0].barh(bottom_5_male_literacy['District name'], bottom_5_male_literacy['Male_Literacy'], color='yellow', label='Bottom 5')
axs[1, 0].set_title('Top and Bottom 5 Male Literacy Rates')
axs[1, 0].set_xlabel('Male Literacy Rate (%)')
axs[1, 0].legend()

# Female Literacy Plot
axs[1, 1].barh(top_5_female_literacy['District name'], top_5_female_literacy['Female_Literacy'], color='pink', label='Top 5')
axs[1, 1].barh(bottom_5_female_literacy['District name'], bottom_5_female_literacy['Female_Literacy'], color='gray', label='Bottom 5')
axs[1, 1].set_title('Top and Bottom 5 Female Literacy Rates')
axs[1, 1].set_xlabel('Female Literacy Rate (%)')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
state_grouped = df_2011.groupby('State name').agg({
    'Male': 'sum',
    'Female': 'sum',
    'Literate': 'sum',
    'Male_Literate': 'sum',
    'Female_Literate': 'sum',
    'Population': 'sum'
}).reset_index()

# Calculate the metrics for each state
state_grouped['Gender_Ratio'] = state_grouped['Male'] / state_grouped['Female']
state_grouped['Literacy_Rate'] = (state_grouped['Literate'] / state_grouped['Population']) * 100
state_grouped['Male_Literacy'] = (state_grouped['Male_Literate'] / state_grouped['Male']) * 100
state_grouped['Female_Literacy'] = (state_grouped['Female_Literate'] / state_grouped['Female']) * 100

# Get top and bottom 5 states for each metric
top_5_state_gender_ratio = state_grouped.nlargest(5, 'Gender_Ratio')
bottom_5_state_gender_ratio = state_grouped.nsmallest(5, 'Gender_Ratio')

top_5_state_literacy_rate = state_grouped.nlargest(5, 'Literacy_Rate')
bottom_5_state_literacy_rate = state_grouped.nsmallest(5, 'Literacy_Rate')

top_5_state_male_literacy = state_grouped.nlargest(5, 'Male_Literacy')
bottom_5_state_male_literacy = state_grouped.nsmallest(5, 'Male_Literacy')

top_5_state_female_literacy = state_grouped.nlargest(5, 'Female_Literacy')
bottom_5_state_female_literacy = state_grouped.nsmallest(5, 'Female_Literacy')

# Plotting the top and bottom 5 for each metric for states

fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Gender Ratio Plot for States
axs[0, 0].barh(top_5_state_gender_ratio['State name'], top_5_state_gender_ratio['Gender_Ratio'], color='green', label='Top 5')
axs[0, 0].barh(bottom_5_state_gender_ratio['State name'], bottom_5_state_gender_ratio['Gender_Ratio'], color='red', label='Bottom 5')
axs[0, 0].set_title('Top and Bottom 5 State Gender Ratios')
axs[0, 0].set_xlabel('Gender Ratio')
axs[0, 0].legend()

# Literacy Rate Plot for States
axs[0, 1].barh(top_5_state_literacy_rate['State name'], top_5_state_literacy_rate['Literacy_Rate'], color='blue', label='Top 5')
axs[0, 1].barh(bottom_5_state_literacy_rate['State name'], bottom_5_state_literacy_rate['Literacy_Rate'], color='orange', label='Bottom 5')
axs[0, 1].set_title('Top and Bottom 5 State Literacy Rates')
axs[0, 1].set_xlabel('Literacy Rate (%)')
axs[0, 1].legend()

# Male Literacy Plot for States
axs[1, 0].barh(top_5_state_male_literacy['State name'], top_5_state_male_literacy['Male_Literacy'], color='purple', label='Top 5')
axs[1, 0].barh(bottom_5_state_male_literacy['State name'], bottom_5_state_male_literacy['Male_Literacy'], color='yellow', label='Bottom 5')
axs[1, 0].set_title('Top and Bottom 5 State Male Literacy Rates')
axs[1, 0].set_xlabel('Male Literacy Rate (%)')
axs[1, 0].legend()

# Female Literacy Plot for States
axs[1, 1].barh(top_5_state_female_literacy['State name'], top_5_state_female_literacy['Female_Literacy'], color='pink', label='Top 5')
axs[1, 1].barh(bottom_5_state_female_literacy['State name'], bottom_5_state_female_literacy['Female_Literacy'], color='gray', label='Bottom 5')
axs[1, 1].set_title('Top and Bottom 5 State Female Literacy Rates')
axs[1, 1].set_xlabel('Female Literacy Rate (%)')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate Pearson correlation between Male and Female Literacy
correlation = state_grouped['Male_Literacy'].corr(state_grouped['Female_Literacy'])

print(f"Pearson correlation between Male and Female Literacy: {correlation}")
```

```python
import matplotlib.pyplot as plt

# Scatter plot for Male vs Female Literacy
plt.figure(figsize=(8, 6))
plt.scatter(state_grouped['Male_Literacy'], state_grouped['Female_Literacy'], alpha=0.6, color='b')
plt.title('Male Literacy vs Female Literacy')
plt.xlabel('Male Literacy Rate (%)')
plt.ylabel('Female Literacy Rate (%)')
plt.show()
```

```python
# Calculate the Female to Male Ratio
state_grouped['Female_to_Male_Ratio'] = state_grouped['Female'] / state_grouped['Male']

# Calculate the Pearson correlation between Female to Male Ratio and Overall Literacy Rate
correlation_female_to_male_vs_literacy = state_grouped['Female_to_Male_Ratio'].corr(state_grouped['Literacy_Rate'])

print(f"Pearson correlation between Female to Male Ratio and Overall Literacy Rate: {correlation_female_to_male_vs_literacy}")
```

```python
# Scatter plot for Female to Male Ratio vs Overall Literacy Rate
plt.figure(figsize=(8, 6))
plt.scatter(state_grouped['Female_to_Male_Ratio'], state_grouped['Literacy_Rate'], alpha=0.6, color='green')
plt.title('Female to Male Ratio vs Overall Literacy Rate')
plt.xlabel('Female to Male Ratio')
plt.ylabel('Overall Literacy Rate (%)')
plt.show()
```
